```python
# Load Libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier  # Perbaiki di sini
from sklearn.model_selection import train_test_split  # Perbaiki di sini
from sklearn import metrics
```

```python
col_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'label']

pima = pd.read_csv("diabetes.csv")
```

```python
pima.head()
```

```python
pima.info()
```

```python
feature_cols = ['Insulin', 'BMI', 'Age', 'Glucose', 'BloodPressure', 'DiabetesPedigreeFunction']
X = pima[feature_cols]
Y = pima.Outcome
```

```python
#Split dataset info training [set and test set]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=1) #70% training 30% testing
```

```python
# Create Decission Tree classifer object
clf = DecisionTreeClassifier()

# Train Decission Tree
clf = clf.fit(X_train, Y_train)

# Predict the response for test dataset
Y_pred = clf.predict(X_test)
```

```python
# Model Acuracy, how often is the classifeier correct
print("Accuracy:", metrics.accuracy_score(Y_test, Y_pred))
```

```python
from sklearn import tree  # Correct import
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5), dpi=500)
tree.plot_tree(clf,
              feature_names=feature_cols,  # Changed to feature_names
              class_names=["0", "1"],
              filled=True,
              rounded=True);
plt.savefig('dtree.png')
```
